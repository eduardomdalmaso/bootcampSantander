## Extração do dataset

```python
import pandas as pd
# ** EXTRACT **
df = pd.read_csv('E:\BootCampSantander\etl\pokePd.csv')
```

```python
# Lendo o cabeçalho do DataSet
df.head()
```

```python
# Removendo coluna Unnamed ** TRANSFORM **
df = df.drop("Unnamed: 0", axis = 1)
```

```python
# Verificando
df.head()
```

```python
# ** TRANSFORM ** 2
# Pegar todos os pokemons com HP > 50
df = df[df["HP"] > 50]
```
